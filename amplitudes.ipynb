```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
baselinefreq = [0.5, 1.0, 1.5, 2.0, 3.0, 4.0]
baselineampin = [10]*len(baselinefreq)
baselineexcout = [0.39, 0.33, 0.23, 0.18, 0.16, 0.107]
```

```python
baselinestifffreq = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0]
baselinestiffampin = [10]*len(baselinestifffreq)
baselinestiffexcout = [0.38, 0.34, 0.26, 0.22, 0.19, 0.14, 0.12, 0.12]
```

```python
plt.plot(baselinefreq, baselineexcout, 'o-', label='flex')
plt.plot(baselinestifffreq, baselinestiffexcout, '+-', label='stiff')
plt.legend()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Tail excursion (L)')
plt.show()
```

```python
eqamps = np.array(baselinestiffampin)/np.array(baselinestiffexcout)*0.3
eqamps
```

```python
tallerfreq = [0.435, 0.869, 1.303, 1.738, 2.173, 2.608, 3.042, 3.477]
tallerampin = [15]*len(tallerfreq)
tallerexcout = [0.548, .388, 0.282, 0.211, 0.178, 0.149, 0.139, 0.125]
```

```python
tallerstifffreq = [0.435, 0.869, 1.303, 1.738, 2.173, 2.608, 3.042, 3.477]
tallerstiffampin = [15]*len(tallerfreq)
tallerstiffexcout = [0.534, 0.426, 0.336, 0.266, 0.220, 0.184, 0.156, 0.135]
```

```python
plt.plot(tallerfreq, tallerexcout, 'o-', label='flex')
plt.plot(tallerstifffreq, tallerstiffexcout, '+-', label='stiff')
plt.legend()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Tail excursion (L)')
plt.show()
```

```python
eqamps = np.array(tallerstiffampin)/np.array(tallerstiffexcout)*0.3
eqamps
```

```python
shorterfreq = [0.59, 1.19, 1.785, 2.380, 2.976, 3.570, 4.166]
shorterampin = [15]*len(tallerfreq)
shorterexcout = [0.486, 0.344, 0.252, 0.191, 0.171, 0.145, 0.124]
```

```python
shorterstifffreq = [0.59, 1.19, 1.785, 2.380, 2.976, 3.570, 4.166]
shorterstiffampin = [15]*len(tallerfreq)
shorterstiffexcout = [0.5176, 0.396, 0.293, 0.237, 0.201, 0.153, 0.141]
```

```python
plt.plot(tallerfreq, tallerexcout, 'o-', label='flex')
plt.plot(tallerstifffreq, tallerstiffexcout, '+-', label='stiff')
plt.legend()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Tail excursion (L)')
plt.show()
```

```python
plt.plot(baselinefreq, np.array(baselineexcout)*1.5, 'ko-', label='baseline, flex')
plt.plot(baselinestifffreq, np.array(baselinestiffexcout)*1.5, 'k+--', label='baseline, stiff')

plt.plot(tallerfreq, np.array(tallerexcout), 'ro-', label='taller, flex')
plt.plot(tallerstifffreq, np.array(tallerstiffexcout)*1.0, 'r+--', label='taller, stiff')

plt.plot(shorterfreq, np.array(shorterexcout), 'bo-', label='shorter, flex')
plt.plot(shorterstifffreq, np.array(shorterstiffexcout)*1.0, 'b+--', label='shorter, stiff')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Tail excursion (L)')

plt.legend()
plt.show()
```
